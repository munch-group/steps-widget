**steps-widget** traces how Python actually evaluates an expression, one
substitution or reduction at a time. Tag any statement in a cell with a trailing
`# PRINT STEPS` comment and the `%%steps` cell magic renders the evaluation trace
as a widget below the cell -- the same trace the `print-steps` command line tool
prints to the terminal for a whole script.

steps-widget's stepper dispatches on CPython 3.9/3.10 bytecode opcode names that
Python 3.11 restructured, so it only installs on **Python 3.9 or 3.10**.

## The `%%steps` cell magic

Importing `steps_widget` registers the `%%steps` cell magic. The cell still runs
normally; every statement tagged `# PRINT STEPS` also gets a trace card below the
cell showing each substitution and reduction step.

````python
import steps_widget
````

````python
%%steps
x = 7
y = 5
z = x * y + 4  # PRINT STEPS
k = z * 42  # PRINT STEPS
````

You can also build a [](`steps_widget.widget.StepsWidget`) directly from a source
string, the same way the cell magic does internally:

````python
from steps_widget import StepsWidget

StepsWidget('''
z = x * y + 4  # PRINT STEPS
''')
````

## The `print-steps` command line tool

Outside of a notebook, `print-steps` traces every `# PRINT STEPS`-tagged line of a
student script and prints the trace to stderr:

```bash
print-steps student_script.py
```

It first runs the script unmodified to confirm it is error-free, then re-runs an
instrumented copy so line numbers and execution order match the original file.
